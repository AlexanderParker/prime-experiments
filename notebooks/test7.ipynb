```python
import math
import sympy

def find_missing_prime(n, upper_bound):
    lower_bound = math.ceil(math.sqrt(n))
   

    numbers = list(range(lower_bound, upper_bound))
    remainders_set = set()

    for i in numbers:
        remainders_set.add(i - (n % i))

    m = 1 if n % 2 == 0 else 2
    checks = 0
    while m in remainders_set:        
        checks += 1
        m += 2
    print (checks)
    primetest = n + m
    is_prime = sympy.isprime(primetest)
    return primetest, is_prime, m, upper_bound

def find_potential_prime(n):

    upper_bound = math.ceil(math.sqrt(n))

    numbers = list(range(2, upper_bound))
    remainders_set = set()

    for i in numbers:
        remainders_set.add(i - (n % i))

    m = 1 if n % 2 == 0 else 2
    while m in remainders_set:
        m += 2
    primetest = n + m    
    is_prime = sympy.isprime(primetest)
    missing_upper_bound = upper_bound
    was_missing = False
    while not is_prime:        
        mssing_primetest, missing_is_prime, missing_m, missing_upper_bound = find_missing_prime(n, missing_upper_bound)
        if missing_is_prime:
            was_missing = True
            return mssing_primetest, missing_is_prime, missing_m, missing_upper_bound, was_missing
        else:
            missing_upper_bound += 1
    return primetest, is_prime, m, upper_bound, was_missing

max_n = 1000  # You can adjust this value as needed
wrong_guesses = {}
progress_step = max_n // 100 if max_n >= 100 else 1
for n in range(1, max_n + 1):
    primetest, is_prime, m, upper_bound, was_missing = find_potential_prime(n)    
    if not is_prime or was_missing:
        wrong_guesses[n] = {'n':n, 'primetest':primetest, 'is_prime':is_prime, 'm':m, 'upper_bound':upper_bound, 'was_missing':was_missing}
    if n % progress_step == 0 or n == max_n:
        print(f"Checked up to n={n} / {max_n}", end='\r', flush=True)
print()

print(f"Total wrong guesses found: {len(wrong_guesses)}")

print (wrong_guesses.items())

print("\nMapping of wrong guesses to correct guesses (first 20):")
for (n, value) in list(wrong_guesses.items())[:20]:    
    print(f"n={n}: wrong_guess={value['primetest']}, wrong_m={value['m']}, wrong_upper_bound={value['upper_bound']}, was_missing={value['was_missing']}")

# write wrong_guesses to a file
with open("wrong_guesses.csv", "w") as f:
    # write header
    f.write("n,wrong_guess,wrong_m,wrong_upper_bound,was_missing\n")
    for (n, value) in wrong_guesses.items():
        f.write(f"{n},{value['primetest']},{value['m']},{value['upper_bound']},{value['was_missing']}\n")
```
